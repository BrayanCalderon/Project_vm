```python
#Crop de Imagenes 


#Funcion para calcular W y H
def area(box):
    x = box[0,1]
    y = box[0,0]
    dx1 = box[0,1]-box[3,1]
    dy1 = box[0,0]-box[3,0]
    dx2 = box[0,1]-box[1,1]
    dy2 = box[0,0]-box[1,0]
    w   = (dx1**2+dy1**2)**(1/2)
    h   = (dx2**2+dy2**2)**(1/2)
    return x,y,w,h
```

```python
def crop_images(rects,rgb,box):
    # Rota las imagenes
    rect = rects[0]
    angle = rect[2]
    rows,cols = rgb.shape[0], rgb.shape[1]
    M = cv2.getRotationMatrix2D((cols/2,rows/2),angle,1)
    img_rot = cv2.warpAffine(rgb,M,(cols,rows))

    # Realiza el recorte a los contornos
    pts = np.int0(cv2.transform(np.array([box]), M))[0]    
    img_crop = img_rot[pts[1][1]:pts[0][1], pts[1][0]:pts[2][0]]
    return img_crop
```

```python

def draw_contour(img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    th, im_gray_th_otsu = cv2.threshold(img_gray, 128, 192, cv2.THRESH_OTSU)
    src_color = img.copy()
    src_crop = img.copy()

    outer_contours = []
    rects = []

    #Obtengo los contornos y la jerarquia a la que pertenecen
    contours, hierarchy = cv2.findContours(im_gray_th_otsu, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

    #Selecciono aquellos contornos que no tienen padres 
    for i in range(len(hierarchy[0])):
        if hierarchy[0,i,3] < 1:
            outer_contours.append(contours[i])
        else:
            continue
    c = max(outer_contours, key = cv2.contourArea)

    
    #for cnt in c:

        #Obtengo el area minima rectangular de cada contorno
    rect = cv2.minAreaRect(c)
        # obtengo representacin en puntos para obtener el area de cada contorno
    box = cv2.boxPoints(rect)
    box = np.int0(box)
    x ,y ,w, h = area(box)
        # Selecciona los contornos que cumplen ciertas condiciones sino las cumple no dibuja el contorno y pasa al sgte
    #if (w*h >= src_color.shape[0]*src_color.shape[1]) or w*h < 1000000 or w*h > 200000000:
        #continue

    epsilon = 0.01*cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, epsilon, True)
    cv2.drawContours(src_color, [approx], 0, (0,0,255), 20)
    rects.append(rect)
        
    stencil = np.zeros(img.shape).astype(img.dtype)
    color = [255, 255, 255]
    cv2.fillPoly(stencil, [approx], color)
    result = cv2.bitwise_and(src_crop, stencil)
        
        
    plt.imshow(src_crop)
    crop_img = crop_images(rects,result,box)
        
    return crop_img

                   
        
```

```python
img = cv2.imread('IMG_2163.JPG')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(draw_contour(img_rgb))
```

```python
path = "Images/"
files = ls1(path)
for idx,file in enumerate(files):
    filea=path + file # archivo original
    #sext es la extensin y sfilename y es el nombre del archivo sin la extensin
    sext, sfilename = splitfilename(file)
    filec=path+"crop/" + "img"+str(idx)+ ".png" # archivo crop
    #print(file)
    img = cv2.imread(filea)
    crop_img = draw_contour(img)
    cv2.imwrite(filec,crop_img)
    

    
```
